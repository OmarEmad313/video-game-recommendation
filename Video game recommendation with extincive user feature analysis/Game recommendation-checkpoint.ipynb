```python
userDataset = pd.read_csv('user_dataset.csv')
```

```python
dfUsers = pd.DataFrame(userDataset)
#renaming features

dfUsers.rename(columns={'Games you love (at least 5 games)': 'game'},
          inplace=True, errors='raise')
```

```python
#droping unuseful columns
dfUsers.drop(['Timestamp', 'Username'], axis = 1, inplace = True) 
```

```python
list(dfUsers.columns)
```

```python
display(dfUsers)
```

```python
#creating the user id column
dfUsers["id"]=dfUsers.index+1
dfUsers = dfUsers[['id'] + [x for x in dfUsers.columns if x != 'id']]
```

```python
dfUsers["game"].unique
```

# The merged dataset

```python
dataset = pd.read_csv('dataset.csv')
```

```python
df= pd.DataFrame(dataset)
```

```python

df.rename(columns={'Where would you place yourself on the scale below in hours per day of gaming? (كم ساعه تلعب في اليوم)': 'daily hours'},
          inplace=True, errors='raise')
df.rename(columns={'favorite genres': 'favorite_genres'},
          inplace=True, errors='raise')
df.rename(columns={'  Platforms': 'platforms'},
          inplace=True, errors='raise')
```

```python
df.describe()
```

```python
df['age'] = df['age'].astype(int)
df['daily hours'] = df['daily hours'].astype(int)
```

```python
df[df.duplicated()]
```

```python
df_melt = df.assign(favorite_genres=df.favorite_genres.str.split(";"))
```

```python
df_melt.drop(['Unnamed: 0', 'id'], axis = 1, inplace = True) 
```

```python
df_melt=df_melt.explode("favorite_genres")
```

```python
df_melt["id"]=df_melt.index+1
df_melt = df_melt[['id'] + [x for x in df_melt.columns if x != 'id']]
```

```python
df_melt = df_melt.assign(platforms=df_melt.platforms.str.split(";"))
```

```python
df_melt=df_melt.explode("platforms")
```

```python
display(df_melt)
```

```python
df=df_melt
```

```python
first_column = df.iloc[:, 0]
```

```python
df=df.reset_index(drop=True)
```
